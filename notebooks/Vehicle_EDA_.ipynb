```python
print('Missing values')
car_prices.isna().sum()
```

Accodring to code above we need to handle fetures which has null values:
In categorical features we will impute 'other' or take mode instead of null values. In numerical features we will impute median or mean instead of null values according to data type and data distribution. Data quality:excelent dataset is almost fullfilled. Dataset have many features.(according to datadensity).  

```python
# Fill missing values with 'other' category
car_prices['trim'] = car_prices['trim'].fillna('other')
car_prices['color'] = car_prices['color'].fillna('other')
car_prices['make'] = car_prices['make'].fillna('other')
car_prices['model'] = car_prices['model'].fillna('other')


# Fill missing values with mode
car_prices['body'] = car_prices['body'].fillna(car_prices['body'].mode()[0])
car_prices['interior'] = car_prices['interior'].fillna(car_prices['interior'].mode()[0])
car_prices['transmission'] = car_prices['transmission'].fillna(car_prices['transmission'].mode()[0])

# Remove null values
car_prices.dropna(subset=['vin'], inplace=True)
car_prices.dropna(subset=['saledate'], inplace=True)

```

```python
car_prices['condition'] = car_prices['condition'].fillna(car_prices['condition'].median())
car_prices['odometer'] = car_prices['odometer'].fillna(car_prices['odometer'].mean())
car_prices['mmr'] = car_prices['mmr'].fillna(car_prices['mmr'].mean())
```

Numerical features needed to be scaled and categorical features needed to be encoded. We use label encoder because there are a lot of unique models of cars.

```python
car_prices.drop(['saledate'],axis=1,inplace=True)
car_prices.head(3)
# We faced the problem with parsing saledate so we decided to drop and use year as datetime object
```

```python
from sklearn.preprocessing import MinMaxScaler, LabelEncoder

def normalize_features(df):
    # Copy the original DataFrame to avoid modifying the input
    normalized_df = df.copy()

    # Numerical columns to be normalized
    numerical_cols = ['condition', 'odometer', 'mmr']

    # Categorical columns to be encoded
    categorical_cols = [col for col in df.columns if col not in numerical_cols and col != 'sellingprice' and col !='year']

    # Normalize numerical features using Min-Max Scaling
    scaler_dict = {}
    for col in numerical_cols:
        scaler = MinMaxScaler()
        normalized_df[col] = scaler.fit_transform(df[[col]])
        scaler_dict[col] = scaler

    # Encode categorical features using Label Encoding
    label_encoders = {}
    for col in categorical_cols:
        le = LabelEncoder()
        normalized_df[col] = le.fit_transform(df[col])
        label_encoders[col] = le

    return normalized_df, scaler_dict, label_encoders
```
